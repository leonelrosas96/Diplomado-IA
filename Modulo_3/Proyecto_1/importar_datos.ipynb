```python
#Eliminar columnas que no brindan informacion

valores_unicos_n = df.nunique()

columnas_sn_info = valores_unicos_n[valores_unicos_n == 1].index.to_list()
columnas_sn_info
```

```python
#Para borrar columnas y valores usar drop
#recordar que el lenguaje para coluna y fila en data frame es (0,1) con axis
#Para reemplazar los datos colocar inplace 

#df.drop(columnas_sn_info,axis=1, inplace= True) <- esto es para eliminar el data frame original y reemplazar con las nuevas indicaciones


#Otra forma de reemplazar pero asignar un valor sin modificar la tabla original 
df_clean = df.drop(columnas_sn_info,axis=1)
df_clean
```
