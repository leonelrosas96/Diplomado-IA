```python
#Graficos de serie de tiempo

import pandas as pd
```

```python
data_range = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
precios = np.cumsum(np.random.randn(len(data_range))) + 100

```

```python
data = pd.DataFrame({'Date': data_range, 'Precios': precios})
data.head()
```

```python
plt.plot(data['Date'],data['Precios'], label = 'Precio')
plt.show()
```

### Ejercicio 

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

categorias = ['A','B','C','D']
valores = [5,7,3,8]

spx = np.random.rand(100)
spy = np.random.rand(100)

val = [np.random.normal (0,std,100) for std in range (1,3)]
```

```python
fig, axs = plt.subplots(2,2, figsize = (15,10))

axs[0,0].plot(x,y1, label = 'Seno')
axs[0,0].plot(x,y2,label= 'Coseno')
axs[0,0].set_title('Gráfico de lineas')
axs[0,0].legend(loc='upper right')
axs[0,0].grid(True)

axs[0,1].bar(categorias, valores)
axs[0,1].set_title('Gráfico de barras')

axs[1,0].scatter(spx,spy, color ="green")
axs[1,0].set_title('Grafico de dispersion')

axs[1,1].boxplot(val)
axs[1,1].set_title('Grafica de cajas y bigotes')
```
